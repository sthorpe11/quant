```python
import datetime
import numpy as np
import pandas as pd

tsdf = pd.read_csv('./Data/WIKI_PRICES_Full.csv')
tsdf['date'] = pd.to_datetime(tsdf['date'])


# Create a list of all symbols
symbol = tsdf['ticker'].tolist()
symbol = list(set(symbol))
symbol.reverse()
symbol.insert(0,'Date')

#count rows of the ticker A
total_file_rows = len(tsdf.index)-1
ifile = 0
max_rows = 0
while (tsdf['ticker'][ifile] == 'A'):
    ifile = ifile+1
    max_rows = max_rows+1

start = tsdf['date'][0]
end = tsdf['data'][max_rows]
print(start,end)
    
# create dataframe
sdf = pd.DataFrame(index=np.arange(max_rows), columns=symbol)

# Add Dates to dataframe using A
ifile = 0
while (tsdf['ticker'][ifile] == 'A'):
    sdf['Date'][ifile] = tsdf['date'][ifile]
    #print(sdf['Date'][ifile])
    ifile = ifile+1

# Add closing prices for symbols to 
ifile = 0
while (ifile < total_file_rows):
    sym = tsdf['ticker'][ifile]
    ndate = tsdf['date'][ifile]
    sdf[sym][sdf['Date']==ndate] = tsdf['close'][ifile]
    ifile = ifile+1
    if (sym != tsdf['ticker'][ifile]):
        print(sym)
        

       
sdf.to_csv('./Data/stockdata.csv', encoding='utf8')
sdf.head()


```

```python
import datetime
import numpy as np
import pandas as pd
import pandas_datareader.data as web

#Get reference
reference_symbol = ['^VIX']
start = '2016-12-1'
end = '2017-3-1'
start_date = pd.to_datetime(start, format='%Y-%m-%d', errors='ignore')
end_date = pd.to_datetime(end, format='%Y-%m-%d', errors='ignore')

sdf = web.DataReader('^GSPC', "yahoo", start_date, end_date)
sdf = sdf.drop(['Open', 'High', 'Low', 'Adj Close', 'Volume'], axis=1)
sdf.rename(columns={0:'Date'},inplace=True)
sdf.rename(columns={'Close':'SP500'},inplace=True)

for sym in reference_symbol:
    tsdf = web.DataReader(sym, "yahoo", start_date, end_date)
    sdf[sym] = tsdf['Close']


sdf.head()
```
